```python
import torch
import captum

from matplotlib import pyplot as plt

from exp_readout import rbf_softeq, relu_softeq
```

```python
x = torch.linspace(-3, 3, steps=int(1e7))
expected = 0.0

y_rbf = rbf_softeq(x, expected)
y_relu = relu_softeq(x, expected)
```

```python
plt.scatter(x, y_rbf, s=0.1)
plt.title("RBF soft equality")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
plt.scatter(x, y_relu, s=0.1)
plt.title("ReLU soft equality")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
